```python
zillow_df = pd.read_csv('../raw_data/zillow_data.csv')
```

```python
zillow_citygroup = zillow_df.groupby(['State', 'City'])
# zillow_citygroup.ngroups
# zillow_citygroup.groups.keys()
```

```python
chicago_df = zillow_citygroup.get_group(('IL', 'Chicago'))
# chicago_df[chicago_df['RegionName']==60657].T.iloc[7:]
chicago_df.head()
```

```python
#time range of data
time_range = chicago_df.columns.tolist()[7:]
```

```python
#long dataframe format
chicago_longdf = chicago_df[time_range].T

#converting index to datetime index
chicago_longdf.index = pd.to_datetime(chicago_longdf.index, format = '%Y-%m')
```

```python
chicago_longdf.head()
```

```python
## chicago county 0
chicago_zero = chicago_longdf[[0]]
chicago_zero.columns = chicago_zero.columns.astype(str)
chicago_zero.columns = ['price']
chicago_zero.head(2)
```

```python
plt.plot(chicago_zero)
plt.title('Chicago county housing prices over 20 year period')
plt.ylabel('Price')
plt.xlabel('Year')
plt.show()
```
